```python
from google import genai
from google.genai import types
import pandas as pd
from google.colab import userdata
```

```python
from google.colab import userdata
api_key = userdata.get('GEMINI_API_KEY')
client = genai.Client(api_key=api_key)
```

```python
# Load Data
holdings_file = client.files.upload(file='/content/holdings.csv')
trades_file = client.files.upload(file='/content/holdings.csv')
```

```python
#context = f"""
#DATASET_1 (Holdings):
#{holdings.to_string(index=False)}

#DATASET_2 (Trades):
#{trades.to_string(index=False)}
#"""
```

```python
def chat_with_data(query):
    system_prompt = (
        "You are a specialized Financial Data Bot. You have access to 'holdings.csv' and 'holdings.csv'. "
        "1. Use the code_execution tool to find answers in these files. "
        "2. IF THE ANSWER IS NOT FOUND IN THE GIVEN FILES, you MUST respond with: 'Sorry can not find the answer'. "
        "3. Do not use any outside knowledge or the internet to answer. "
        "4. Do not provide explanations if the data is missing; only use the required phrase."
    )

    # We define the tool as a dictionary to avoid the AttributeError
    tools_config = [{'code_execution': {}}]

    response = client.models.generate_content(
        model="gemini-2.5-flash",
        config=types.GenerateContentConfig(
            system_instruction=system_prompt,
            tools=tools_config,
            temperature=0.0
        ),
        contents=[holdings_file, trades_file, query]
    )
    return response.text
```

```python
# 3. Test the Bot 1
print("--- Financial Analysis ---")
try:
    print(chat_with_data("Which funds performed better based on yearly P/L? Summarize the results."))
except Exception as e:
    print(f"Error: {e}")
```

```python
# 3. Test the Bot 2
print("--- Financial Analysis ---")
try:
    print(chat_with_data("Compare the total trade volume in trades.csv with the final holdings in holdings.csv for the Ytum fund."))
except Exception as e:
    print(f"Error: {e}")
```
